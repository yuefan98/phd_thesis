# Figure 2. 
### Results from Equation 6 (Imaginary to Real)

```python
df_Z1 = pd.read_csv('data/KK_results/Z1_KK.csv')
df_Z2 = pd.read_csv('data/KK_results/Z2_KK.csv')
```

```python
Z1 = df_Z1['Z1_real']+1j*df_Z1['Z1_imag']
Z1_KK = df_Z1['Z1_KK_real']+1j*df_Z1['Z1_imag']
Z2 = df_Z2['Z2_real']+1j*df_Z2['Z2_imag']
Z2_KK = df_Z2['Z2_KK_real']+1j*df_Z2['Z2_imag']
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4.7))
for idx, axi in enumerate(ax.flat):    
    axi.text(-0.21, 1, f'({string.ascii_lowercase[idx]})',horizontalalignment='center',verticalalignment='center', transform=axi.transAxes, size=22, weight='bold')

plot_first(ax[0],Z1,fmt ='d', ms = 10)
plot_first(ax[0],Z1_KK, fmt ='o', ms = 5)

plot_second(ax[1],Z2,fmt ='d', ms = 10)
plot_second(ax[1],Z2_KK,fmt ='o', ms = 5)
ax[0].legend(['Linear RC', 'Kramers-Kronig Transform'],fontsize = 14)
ax[1].legend(['Weakly Nonlinear RC', 'Kramers-Kronig Transform'],fontsize = 14)
plt.tight_layout()
plt.show()
```

# Figure S2.
### Results from Equation 7 (Real to Imaginary)

```python
df_Z1_Re_to_Im = pd.read_csv('data/KK_results/Z1_KK_Re_to_Im.csv')
df_Z2_Re_to_Im = pd.read_csv('data/KK_results/Z2_KK_Re_to_Im.csv')
```

```python
Z1_KK_Re_to_Im = df_Z1['Z1_real']+1j*df_Z1_Re_to_Im['Z1_KK_imag']
Z2_KK_Re_to_Im = df_Z2['Z2_real']+1j*df_Z2_Re_to_Im['Z2_KK_imag']
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4.7))
for idx, axi in enumerate(ax.flat):    
    axi.text(-0.21, 1, f'({string.ascii_lowercase[idx]})',horizontalalignment='center',verticalalignment='center', transform=axi.transAxes, size=22, weight='bold')

plot_first(ax[0],Z1,fmt ='d', ms = 10)
plot_first(ax[0],Z1_KK_Re_to_Im, fmt ='o', ms = 5)

plot_second(ax[1],Z2,fmt ='d', ms = 10)
plot_second(ax[1],Z2_KK_Re_to_Im,fmt ='o', ms = 5)
ax[0].legend(['Linear RC', 'Kramers-Kronig Transform'],fontsize = 14)
ax[1].legend(['Weakly Nonlinear RC', 'Kramers-Kronig Transform'],fontsize = 14)
plt.tight_layout()
plt.show()
```
