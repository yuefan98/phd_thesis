```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('poly_1st_c.csv',names =['x','y'])
```

```python
d111 = 2.41217 
y111 = d111*6
```

```python
np.arange(3.6,4.6,0.2)
```

```python
fig,ax = plt.subplots(figsize = (5,5))
ax.plot(df['x'],100*(df['y']-y111)/y111,'-o' )
ax.set_xlabel('Potential $U^{+}$, [V vs Li/Li$^+$]',fontsize = 16)
ax.set_ylabel('Lattice Strain, [%]',fontsize = 16)
ax.tick_params(axis='x', which='major', labelsize=14) 
ax.tick_params(axis='y', which='major', labelsize=14) 
ax.set_xticks(np.arange(3.6,4.6,0.2))
ax.set_yticks(np.arange(-2,0.5,0.5))
plt.show()
# fig.savefig('figures/lattice_strain.png',dpi=300,bbox_inches='tight')
```
